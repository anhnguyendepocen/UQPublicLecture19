```python
import numpy as np
from numpy.random import randn, seed
import matplotlib.pyplot as plt

seed(4)
n = 10000
h = 1/n
W = np.cumsum(randn(n,2)*np.sqrt(h),0)
plt.plot(W[:,0],W[:,1])


for i in range(2):
    Z = randn(n,2)
    X = np.zeros((n,2))
    t = np.arange(n)/n
    xend = np.array([4,4])
    xstart = np.array([0,0])
    for k in range(1,n):
        mu = X[k-1] +  (xend - X[k-1])*(t[k]-t[k-1])/(1- t[k-1])
        sig2 = (1 - t[k])*(t[k]-t[k-1])/(1 - t[k-1])
        X[k]= mu + np.sqrt(sig2)*Z[k-1]
        
    plt.plot(X[:,0],X[:,1])
    plt.tight_layout()
    plt.savefig("bbridge")
```
