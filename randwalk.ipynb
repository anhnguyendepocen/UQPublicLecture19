```python
from numpy.random import rand
import numpy as np
import matplotlib.pyplot as plt
n =1000
p = 0.5
u = rand(n)
x = (u<p).astype(int)
y = 2*x - 1
y = np.concatenate(([0],y))
z = np.cumsum(y)
plt.plot(np.arange(n+1),z)
plt.show()
```
